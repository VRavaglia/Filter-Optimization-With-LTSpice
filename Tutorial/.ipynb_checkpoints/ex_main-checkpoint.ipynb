Os parâmetros a serem otimizados são: W de M1 (W1), W de M2 (W2), L de M1 (L1), L de M2 (L2) e Vcomp

A simulação realizada é um DC sweep em Vin. 

Idealmente, para valores de Vin menores do que o limiar desejado de comparação,

a saída Vout deve valer Vdd, e deve valer 0 caso contrário

## Algoritmo Genético em Si

```python
# Contabilizar tempo necessario
st = time.time()

# circ_name: nome do arquivo com o circuito a ser otimizado
# runs: numero de rodadas de otimizacao independentes
# n_params: numero de parametros a serem otimizados
# mu: numero de individuos por geração
# gens: numero total de gerações
# mut_rate: taxa de mutação
# k: numero de individuos que servirão de possiveis pais
# target: tensão alvo do comparador

circ_name = "comparator"
runs = 8
n_params = 5
mu = 10
gens = 15
mut_rate = 0.2
k = 2
target = 1

# Fixar a seed para manter resultados consistentes
np.random.seed(0)

# Parametros da mutação e da inicialização
# Em ordem: W1, W2, L1, L2, Vcomp
# granularity: minimo incremento na mutação
# max_inc: maximo incremento na mutação
# limits:  valores maximos e minimos apos a mutação
# Apenas inteiros são suportados

granularity = np.array([50, 50, 50, 50, 100])
max_inc = np.array([200, 200, 200, 200, 300])
limits = np.array([[500, 5000], [500, 5000], [500, 5000], [500, 5000], [0, 1800]])

# Inicialização aleatoria dos parametros dentro dos limites
# Xs: "mu" individuos com "n_params" parametros para "runs" rodadas

Xs = np.zeros((mu, n_params,runs))
for i in range(n_params):
    Xs[:, i, :] = np.random.randint(limits[i, 1] - limits[i, 0], size=(mu, runs)) + limits[i, 0]

# Xs_mult: multiplicador dos parametros e.g. W de 1u = 1000 * 1e-9 ou 1 * 1e-6
# Xs_names: nome das variaveis no arquivo de simulação que serão otimizadas
# a ordem deve ser a mesma em todos os parametros de mutação e inicialização
    
Xs_mult = np.array([1e-9, 1e-9, 1e-9, 1e-9, 1e-3])
Xs_names = np.array(["W1", "W2", "L1", "L2", "Vcomp"])

# Chamada do algoritmo genetico em si
# bestS: melhor fitness dentre todas as geraçoes, para cada rodada
# bestXs: melhor individuo dentre todas as geraçoes, para cada rodada
# Jhist: historico de todas as fitness para cada rodada
bestS, bestXs, Jhist = SGA (Xs, Xs_mult, Xs_names, gens, k, mu, max_inc, n_params, limits, granularity, mut_rate, runs, circ_name, target)

# Calculo do tempo necessario
et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')
```

## Plot do historico de aptidões

```python
for i in range(0, runs):
    plt.plot(Jhist[:, i])
plt.xlabel("Generation")
plt.ylabel("Fitness Score")
plt.grid()
plt.show()
```

## Cria um arquivo e simula o melhor individuo de todos

```python
# Acha o melhor individuo baseando-se em quem tem o menor fitness
bestX = bestXs[:, np.argmin(bestS)]
print("BestX: ", bestX)

# Cria um esquematico com apenas 1 individuo para ser simulado
# O esquematico representa a rodada "99" para evitar conflitos com outros arquivos
lt_create_gen(circ_name, bestX, Xs_mult, Xs_names, 99)

# Simula a rodada "99" (melhor individuo)
results = lt_simulate(circ_name, 99)

if results.wait() != 0:
    print("There was an error")

# Le e interpreta os resultados da simulação do melhor individuo
sim_type = 'dc'
out_names = ["V(Vout)"]
    
bestSweep, bestOuts = lt_parse(circ_name, sim_type, 99, out_names)
final_outs = []
for j in range(len(out_names)):
    final_out = bestOuts[0,j, :]
    final_outs.append(final_out)
```

## Plot do resultado final

```python
plt.plot(bestSweep[0], final_outs[0], label="Best Individual")
target_signal = (bestSweep[0] < target)*1.8
plt.plot(bestSweep[0], target_signal, '--', label="Ideal Output")

plt.legend(loc="upper right")

plt.show()
print("Best score: ", np.min(bestS))
```

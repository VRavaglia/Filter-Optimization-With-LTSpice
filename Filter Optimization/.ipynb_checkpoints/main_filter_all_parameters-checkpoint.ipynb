```python
import matplotlib.pyplot as plt
import time
```

```python
%run ./ltspice_helper.ipynb
%run ./SGA.ipynb
```

```python
circ_name = "filter_optimized_mosfet_multiple_umc_all"
runs = 8
n_params = 15
mu = 50
gens = 50
mut_rate = 0.3
k = 4
target = 4e6
heuristic = True

np.random.seed(0)

max_inc = np.concatenate((np.array([200]*6),
                          np.array([2]*6),
                          np.array([20]*3)))

limits = np.concatenate((np.array([[240, 5000]]*6),
                         np.array([[1, 5]]*6),
                         np.array([[10, 100]]*3)))

granularity = np.concatenate((np.array([50]*6),
              np.array([1]*6),
              np.array([5]*6)))
```

```python
st = time.time()

Xs = np.zeros((mu, n_params,runs))

for i in range(n_params):
    Xs[:, i, :] = np.random.randint(limits[i, 1] - limits[i, 0], size=(mu, runs)) + limits[i, 0]
    
Xs_mult = np.concatenate((np.array([1e-9]*6),
                          np.array([1]*6),
                          np.array([1e-13]*3)))

Xs_names = np.array(["W1", "W2", "W3", "W4", "W5", "W6", "M1", "M2", "M3", "M4", "M5", "M6", "C1", "C2", "C3"])

bestS, bestXs, Jhist = SGA (Xs, Xs_mult, Xs_names, gens, k, mu, max_inc, n_params, limits, granularity, mut_rate, runs, circ_name, target, heuristic)

et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 4), dpi=80)

for i in range(0, runs):
    plt.plot(Jhist[:, i])
plt.xlabel("Generation", fontsize=16)
plt.ylabel("Fitness Score", fontsize=16)
plt.grid()

plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.locator_params(axis='both', nbins=6)

plt.show()
```

```python
bestX = bestXs[:, np.argmin(bestS)]
print("BestX: ", bestX)
lt_create_gen(circ_name, bestX, Xs_mult, Xs_names, 999)

results = lt_simulate(circ_name, 999)

if results.wait() != 0:
    print("There was an error")

sim_type = 'ac'
out_names = [["V(vo1+)", "V(vo1-)"], ["V(vo2+)", "V(vo2-)"], ["V(vo3+)", "V(vo3-)"], ["V(vo4+)", "V(vo4-)"], ["V(vo5+)", "V(vo5-)"]]
    
bestTime, bestOuts = lt_parse(circ_name, sim_type, 999, out_names)
final_outs = []
for j in range(len(out_names)):
    final_out = bestOuts[0,j, 0, :] + bestOuts[0,j, 1, :]
    final_outs.append(final_out)
```

```python
figure(figsize=(8, 4), dpi=80)

for i in range(5):
    plt.plot(bestTime[0], 20*np.log10(final_outs[i]), label="Vctrl = "+str(0.6 + i*0.05))
    
b, a = signal.cheby1(3, 3, 2*np.pi*target, 'low', analog=True)
_, h = signal.freqs(b, a, bestTime[0]*2*np.pi)
target_signal = 20*np.log10(abs(h))
plt.semilogx(bestTime[0], target_signal, '-.', linewidth='4', label='Target (Vctrl = 0.7)')

plt.plot([target, target], [50, -1000], '--', label='Cutoff = ' + str(np.round(lt_cutoff(bestTime[0], final_outs[2])/1e6, 2))+' MHz')
plt.legend(loc="lower left", prop={'size': 11})
plt.ylim([-10, 2])
plt.xlim([1e5, 1e7])
plt.xscale('log')

plt.xlabel("Frequency (Hz)", fontsize=16)
plt.ylabel("Amplitude (dB)", fontsize=16)
plt.grid()

plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.grid()
plt.show()

print("Best cutoff:", lt_cutoff(bestTime[0], final_outs[2]))
print("Best score: ", np.min(bestS))

```
